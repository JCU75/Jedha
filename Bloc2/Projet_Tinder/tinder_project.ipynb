### Assumptions about causes of missing data of columns "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"
Missing data can be explained by several factors:
* **Lack of participant attendance:** Participants like participant with id 28, did not answer any of the columns, maybe because they did not know or did not want to.
* **Quality rating:** Participants had 100 points to distribute among all the qualities. However, when they had already reached the total of 100 points, they did not bother to put 0 in the remaining columns.

**For columns "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1" :** 
- when the sum is equal to ``100``, fill other columns with ``0`` => exemple if attr1_1 = 50, sinc1_1 = 40, intel1_1 = 10, amb1_1 = **NaN**, shar1_1 = **NaN** then =>  amb1_1 = **0**, shar1_1 = **0**.

```python
def fill_with_zero_when_100_reached(row):
    normalized_row = row.copy()

    total = row[df_signup_qualities_partner].sum()
    
    # if total row under 100, return the row
    if total < 100:
        return normalized_row
  
    if total >= 100:
        cumulative_sum = 0
        for i, col in enumerate(df_signup_qualities_partner):
            cumulative_sum += row[col]
            if cumulative_sum >= 100:
                # replace NaN with 0
                normalized_row.loc[df_signup_qualities_partner[i+1:]] = 0
                break
    
    return normalized_row
```

```python
df[df_signup_qualities_partner] = df.apply(fill_with_zero_when_100_reached, axis=1)[df_signup_qualities_partner]
```

```python
for column_name in df_signup_qualities_partner:
    isna_count = df[column_name].isna().mean() * 100
    print(f"{column_name}: {(isna_count):.2f}%")
```

### The remaining missing values â€‹â€‹come from participants (28, 58, 59, 136, 339, 340, 346) who did not answer all survey questions for columns "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1", but still answered survey questions for other columns, so we cannot remove them.

### Assumptions about causes of missing data of columns "attr", "sinc", "intel", "fun", "amb", "shar"
Missing data can be explained by several factors:
* **Quality rating:** Here, survey participants must rate each quality out of 10. We can therefore assume that for a participant, if he has filled in certain columns and not others, it is because he has not bothered to put 0

**For columns "attr", "sinc", "intel", "fun", "amb", "shar" :** 
- if one of the columns is filled, then the other columns that contain NaN are replaced by 0 => exemple if attr = 5, sinc = 4, intel = 10, amb = **NaN**, shar = **NaN** then =>  amb = **0**, shar = **0**.

```python
def replace_nan_with_zero(row):
   if row.dropna().any():
       # Replace NaN with 0
       # Adding .infer_objects(copy=False) to avoid copying the data and avoid warnings
       return row.fillna(0).infer_objects(copy=False)
   return row
```

```python
df[df_scorecard_qualities_partner] = df.apply(replace_nan_with_zero, axis=1)[df_scorecard_qualities_partner]
```

```python
for column_name in df_scorecard_qualities_partner:
    isna_count = df[column_name].isna().mean() * 100
    print(f"{column_name}: {(isna_count):.2f}%")
```

**We need to normalize the values â€‹â€‹of the scores: ** 
- Indeed before the speed dating, the participant has 100 points to distribute between all the qualities. 
- During the speed dating the participant must rate each quality out of 10.

For the quality scores during speed dating, we need to find a way to have a distribution logic out of 100. To do this we will divide each quality score by the sum of all the scores and then multiply it by 100.
Example for a row:
- attr: 8 => 8/38 * 100 = 21.05
- sinc: 7 => 7/38 * 100 = 18.42
- intel: 8 => 8/38 * 100 = 21.05
- fun: 5 => 5/38 * 100 = 13.16
- amb: 4 => 4/38 * 100 = 10.53
- shar: 6 => 6/38 * 100 = 15.79
- total = 38 
- new_total = 21.05 + 18.42 + 21.05 + 13.16 + 10.53 + 15.79 = 100


```python
df['sum_qualities_during_date'] = df[df_scorecard_qualities_partner].sum(axis=1)
print(df['sum_qualities_during_date'])
```

```python
# normalize score during speed dating for columns "attr", "sinc", "intel", "fun", "amb", "shar"
for col in df_scorecard_qualities_partner:
    df[col + "_norm"] = ((df[col]  / df['sum_qualities_during_date']) * 100).round(2)
```

```python
columns_of_interest_before_date = df_signup_qualities_partner + ["dec"]
columns_of_interest_during_date = df_scorecard_qualities_partner_norm + ["dec"]
```

```python
df_men = df[df['gender']== 1 ]
df_women = df[df['gender']== 0 ]   
```

```python
# select color base on Telegram color to avoid stereotype
men_color = "rgb(0, 196, 170)"
women_color = "rgb(135, 0, 249)"
```

```python
def analyze_correlations_by_gender(df_men, df_women, columns, target_column):
    corr_men = df_men[columns].corr()[target_column].drop(target_column)
    corr_women = df_women[columns].corr()[target_column].drop(target_column)
    
    comparison_df = pd.DataFrame({
        'Men': corr_men,
        'Women': corr_women
    })
    return comparison_df
```

```python
comparison_df_before = analyze_correlations_by_gender(df_men, df_women, columns_of_interest_before_date, "dec")
print("Comparison of correlations by gender before date:")
print(comparison_df_before)
```

```python
comparison_df_during = analyze_correlations_by_gender(df_men, df_women, columns_of_interest_during_date, "dec")
print("Comparison of correlations by gender during the date::")
print(comparison_df_during)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2, subplot_titles=("Before date", "During date"))

# First chart (Before date)
fig.add_trace(go.Bar(x=comparison_df_before.index, y=comparison_df_before['Men'], name="Men",
                    marker_color=men_color),
              row=1, col=1)
fig.add_trace(go.Bar(x=comparison_df_before.index, y=comparison_df_before['Women'], name="Women",
                    marker_color=women_color),
              row=1, col=1)

# Second chart (During date)
fig.add_trace(go.Bar(x=comparison_df_during.index, y=comparison_df_during['Men'], name="Men",
                    marker_color=men_color),
              row=1, col=2)
fig.add_trace(go.Bar(x=comparison_df_during.index, y=comparison_df_during['Women'], name="Women",
                    marker_color=women_color),
              row=1, col=2)



# Personnalisations supplÃ©mentaires
fig.update_layout(
    height=600, width=1200,
    yaxis_title="Coefficient de corrÃ©lation"
)

fig.show()
```
